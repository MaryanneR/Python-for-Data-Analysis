```python
#create histogram off of column A in df1

df1['A'].hist(bins=30)
```

```python
#call column of data frame

df1['A'].plot(kind='hist',bins=30)
```

```python
# area plot

df2.plot.area(alpha=.4)
```

```python
# Bar Plot
#can make stacked by passing in stacked=True

df2.plot.bar(stacked=True)
```

```python
# Histogram

df1['A'].plot.hist(bins=50)
```

```python
# Line plots
#have to y

df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
# Scatter plots
# can use another column to set color of scatterplot
# can also use another column to set size instead of color s=column*sizefactor

# df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')

df1.plot.scatter(x='A',y='B',s=df1['C']*70)
```

```python
# Box Plot

df2.plot.box()
```
