```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#functional method of creating a plot
plt.plot(x,y)
#if not in jupyter notebook, include plt.show() on this line to see the created plot

#adding x and y labels and title to plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#multiplots on same canvas
```

```python
#parameters (# of rows, # of columns, plot #)
plt.subplot(1,2,1)
#'r' stands for red color
plt.plot(x,y,'r')

plt.subplot(1,2,2)
#'b' stands for blue color
plt.plot(y,x,'b')

#can also make plots on one line
#fig,axes = plt.subplots(1,2)
```

```python
#OO - Object Oriented Method
fig = plt.figure() #blank canvas
#more manual method of plot creation

#pass in list with 4 arguments: left, bottom, width, height - range from 0 to 1
axes = fig.add_axes([0.1,0.1,0.8,0.8])

#plot on axes
axes.plot(x,y)

#add labels
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
#subplots
#does add_axes for you based on number of rows and columns, unline in .figure()
fig,axes = plt.subplots(nrows=1,ncols=2)

#for current_ax in axes:
    #current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

#takes care of any overlapping plots
plt.tight_layout()
```

```python
#Figure Size and DPI
```

```python
#plt.figure(figsize(x,y),dpi=x)
#figsize (width,height in inches)
#dpi - dots/pixels per inch

fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label='X Cubed')

#adding legend
#need to include label parameter in plot axes in order for ax.legend() to work
#can specify location of legend using loc=x parameter, being a location code
#can also add in a 2-tuple x,y of the lower-left corner of the legend ax.legend(loc=(x,y))
ax.legend()
```

```python
#setting colors
#colors parameter can take simple string arguments e.g. 'blue' 'green' etc.
#can also take RGB hex codes e.g. '#FF8C00'
#default linewidth is 1
#linewidth shorthand is lw
#alpha - controls line transparency
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=3,alpha=0.5)
```

```python
#linestyle - e.g.dash line, dotted line etc.
#linestyle shorthand - ls
# '--' is dashed line
# '-.' dash point line
# ':' - dotted line
# 'steps' - stepped line
# '-' normal line - default value
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=3,ls='--')
```

```python
#marking points on a line
#'o' - dots mark points
#many types of marker codes
#markersize controls mark size
#can also specify marker face color, marker edge width, and marker edge color

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=1,ls='--',marker='o',markersize=5,
       markerfacecolor='blue', markeredgewidth=1, markeredgecolor='black')
```

```python
#Control over axis appearance and plot range
```

```python
#to set axis limit - pass in list with lower bound,upper bound
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
